```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Function for creating Empirical Cummulative Density Function
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y

def bootstrap_replicate_1d(data, func):
    """Generate bootstrap replicate of 1D data."""
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)

#for drawing bootstrap replicates
def draw_bs_replicate(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = func(np.random.choice(data, size=len(data)))

    return bs_replicates
```

# Counter Analysis

It appear that the main aim of this study is to compare the difference in the mean excess in readmision ratio  in hospitals with different level of discharges. In order to determine the right dichotomy for number of discharges, the spead of the data should be explored.

```python
plt.boxplot(clean_hospital_read_df['Number of Discharges'])
plt.show()
```

The boxplot above shows a lot of outliers, hence it will be recreated without outliers.

```python
plt.boxplot(clean_hospital_read_df['Number of Discharges'], 0, '')
plt.show()
```

```python
clean_hospital_read_df['Number of Discharges'].describe()
```

The boxplot and other statistics shown above show that the average number of discharges is 362, but the median is 280 discharges because of the outliers. Therefore, it might be safe to split the data into those have 300 discharges or less and greater that 300.

```python
df300p = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > 300]
df300 = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] <= 300]

df30 = df300['Excess Readmission Ratio'].dropna()
df30p = df300p['Excess Readmission Ratio'].dropna()


print('Number of Hospitals with <= 300 discharges: ' + str(len(df30)))
print('Number of Hospitals with > 300 discharges: ' + str(len(df30p)))
```

This looks like a reasonable subdivision of the data and it will inform rest of the analysis.

Thus, the right hypothesis should check if there us a significant difference in the mean excess readmission ratio  between hospitals with <= 300 discharges and > 300 discharges

$H_0$: $MU300$ - $MU300+$ = 0 

$H_A$: |$MU300$  - $MU300+$| <> 0

Before testing this hypotesis we compare the distribution of the empirical cummulative functions of the excess rate of readmissions for the 2 categories

```python
#Calculate means
mean300 = np.mean(df30)
mean300p = np.mean(df30p)

#Calculate stds
std300 = np.std(df30)
std300p = np.std(df30p)


#Create an empirical sample with the same mean and standard dev as the data

#Create ECDF for males and the bootstapped data
x_emp300,y_emp300 = ecdf(df30)

#Create ECDF for female and the bootstapped data
x_emp300p,y_emp300p = ecdf(df30p )

_ = plt.figure(figsize=(9,5))
# Plot the ECDFs and show the plot
_ = plt.plot(x_emp300, y_emp300, marker='.', linestyle='none', label='Excessive Readmission Ration for <= 300 - Actual data')

_ = plt.plot(x_emp300p, y_emp300p, marker='.', linestyle='none', label='Excessive Readmission Ratio for > 300 - Actual data')
_ = plt.xlabel('Excessive Readmission Ratio')
_ = plt.ylabel('CDF')
_ = plt.legend(['Large Hospitals (> 300 discharges)', 'Small Hospitals (<= 300 discharges)'], fontsize=14)
plt.show()
```

```python
x_large = [a for a in df300p['Number of Discharges']]
y_large = list(df300p['Excess Readmission Ratio'])

x_small = [a for a in df300['Number of Discharges']]
y_small = list(df300['Excess Readmission Ratio'])

fig, ax = plt.subplots(figsize=(12,8))
_ = ax.scatter(x_large, y_large, color='green', alpha=0.1)
_ = ax.scatter(x_small, y_small, color='red', alpha=0.1)


_ = ax.set_xlim([20, 7000])
_ = ax.set_xscale('log')
_ = ax.set_xlabel('Number of discharges (log scale)', fontsize=14)
_ = ax.set_ylabel('Excess rate of readmissions', fontsize=14)
_ = ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=16)
_ = ax.legend(['Large Hospitals (> 300 discharges)', 'Small Hospitals (<= 300 discharges)'], fontsize=14)

ax.grid(True)
fig.tight_layout()
```

The intial exploratory data analysis above indicate that the the  distribution of the excessive readmission ratio for the small and large hospitals are not the same although they appear to be close as shown in the scatter plot. Thus, the hypothesis is necessary to determine the actual proportions. 

```python
from scipy import stats

stats.ttest_ind(df30, df30p, equal_var = False)
```

The p-value is less that 0.1 which mean that we can reject the null hypothesis that the average excessive readmission ratio for small hospitals (<=300 discharges) is the same as that of large hospitals (> 300 discharges). Therefore, it can be concluded that there is a significant difference in the excessive readmission ratio of hospital with lesser discharge and those of higher discharge. Considering that the mean excessive readmission ratio is higher in hospitals with lower discharge ratio, there is a chance that these hospitals gave significantly higher readmission rates.

The visualization presented in the priliminary analysis does not show difference in the two samples. However, the plots shown in the exploratory analysis above show the difference in the 2 samples.
